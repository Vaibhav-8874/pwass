```python
#Theoretical Questions

#1. What is unsupervised learning in the context of machine learning?
#Answer-A type of ML where the model learns patterns from unlabeled data without target output.

#2.How does K-Means clustering algorithm work?
#Answer-It groups data into K clusters by assigning points to nearest centroid and updating centroids iteratively.

#3. Explain the concept of a dendrogram in hierarchical clustering?
#Answer-A tree diagram showing how clusters merge at different distances in hierarchical clustering.

#4. What is the main difference between K-Means and Hierarchical Clustering?
#Answer-K-Means needs K beforehand; hierarchical does not and creates a tree structure.

#5. What are the advantages of DBSCAN over K-Means?
#Answer-Detects noise, handles arbitrary shapes, and does not require number of clusters.

#6. When would you use Silhouette Score in clustering?
#Answer-Used to evaluate clustering quality and choose optimal number of clusters.

#7. What are the limitations of Hierarchical Clustering?
#Answer-Slow, high memory usage, and not suitable for large datasets.

#8. Why is feature scaling important in clustering algorithms like K-Means?
#Answer-Ensures all features contribute equally since clustering uses distance.

#9. How does DBSCAN identify noise points?
#Answer-Points with insufficient neighbors within eps distance are labeled as noise.

#10. Define inertia in the context of K-Means? 
#Answer-Sum of squared distances between points and their cluster centroids.

#11. What is the elbow method in K-Means clustering?
#Answer-Finds optimal K where inertia stops decreasing sharply.

#12. Describe the concept of "density" in DBSCAN?
#Answer-Number of points within a specified radius (eps).

#13. Can hierarchical clustering be used on categorical data?
#Answer-Yes, using distance metrics like Hamming or Jaccard.

#14. What does a negative Silhouette Score indicate? 
#Answer-Indicates incorrect clustering; point is closer to another cluster.

#15. Explain the term "linkage criteria" in hierarchical clustering?
#Answer-Method to measure distance between clusters (single, complete, average, ward).

#16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
#Answer-Because it assumes clusters are equal size and spherical.

#17. What are the core parameters in DBSCAN, and how do they influence clustering? 
#Answer-eps (radius) and min_samples (minimum points to form cluster).

#18. How does K-Means++ improve upon standard K-Means initialization?
#Answer-Chooses better initial centroids to improve accuracy and convergence.

#19. What is agglomerative clustering?
#Answer-Bottom-up clustering where points merge step by step into clusters.

#20. What makes Silhouette Score a better metric than just inertia for model evaluation?
#Answer-It measures both cluster separation and compactness, unlike inertia.
```

```python
#Practical Questions:
#21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot? 
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            marker='X', s=200)
plt.title("K-Means Clustering with 4 Centers")
plt.show()
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart?
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, random_state=42)

scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

plt.bar(k_values, scores)
plt.title("Silhouette Scores")
plt.xlabel("K")
plt.ylabel("Score")
plt.show()
```

```python
#34.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot?
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = data.data

inertia = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(k_values, inertia, marker='o')
plt.title("Inertia vs K")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
#36. Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding noise)
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Load dataset
wine = load_wine()
X = wine.data

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=2, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (exclude noise = -1)
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)

print("Number of clusters (excluding noise):", num_clusters)
```
